```python
import numpy as np
import pandas as pd
```

```python
ratings_data=pd.read_csv('C:/Users/xiaofeng.li/Google Drive/Work Scrip Code/Data/ratings_xf.csv')
ratings_data.head()
```

```python
R_df=ratings_data.pivot_table(index='userID',columns='subgameID',values='rating').fillna(0)
R_df.index
```

```python
R_df.shape
```

```python
R=R_df.as_matrix()
user_ratings_mean=np.mean(R,axis=1)
R_demeaned=R-user_ratings_mean.reshape(-1,1)
```

```python
from scipy.sparse.linalg import svds
U,sigma,Vt=svds(R_demeaned,k=50)
```

```python
sigma=np.diag(sigma)
```

```python
U,sigma,Vt
#U是指每一个user的重要性，数值越大越祥光，
#Vt是指每一个game的重要性，数值越大越相关，
#sigma是指user和game的相关性，
```

```python
all_user_predicted_ratings=np.dot(np.dot(U,sigma),Vt)+user_ratings_mean.reshape(-1,1)
preds_df=pd.DataFrame(all_user_predicted_ratings,columns=R_df.columns)
```

```python
preds_df.index=R_df.index
```

```python
preds_df=preds_df.reset_index()
```

```python
preds_df.head()
```

```python
def recommend_games(predictions_df, userID, subgame_df, original_ratings_df, num_recommendations=5):
    
    # Get and sort the user's predictions
    user_row_number = preds_df[preds_df['userID']==(userID)].index[0] # UserID starts at 1, not 0
    sorted_user_predictions = predictions_df.iloc[user_row_number].sort_values(ascending=False)
    
    # Get the user's data and merge in the movie information.
    user_data = original_ratings_df[original_ratings_df.userID == (userID)]
    user_full = (user_data.merge(subgame_df, how = 'left', left_on = 'subgameID', right_on = 'subgameID').
                     sort_values(['rating'], ascending=False)
                 )

    print ('User {0} has already rated {1} movies.'.format(userID, user_full.shape[0]))
    print ('Recommending the highest {0} predicted ratings movies not already rated.'.format(num_recommendations))
    
    # Recommend the highest predicted rating movies that the user hasn't seen yet.
    recommendations = (subgame_df[~subgame_df['subgameID'].isin(user_full['subgameID'])].
         merge(pd.DataFrame(sorted_user_predictions).reset_index(), how = 'left',
               left_on = 'subgameID',
               right_on = 'subgameID').
         rename(columns = {user_row_number: 'Predictions'}).
         sort_values('Predictions', ascending = False).
                       iloc[:num_recommendations, :-1]
                      )

    return user_full, recommendations
```

```python
games_list=pd.read_csv('C:/Users/xiaofeng.li/Google Drive/Work Scrip Code/Data/ratings_xf.csv')
games_list=pd.DataFrame(data=games_list,columns=['subgameID','subgamename'])
games_list=games_list.drop_duplicates()
games_list.head()
```

```python
#recommend_movies(predictions_df, userID, subgame_df, original_ratings_df, num_recommendations=5
already_rated, predictions = recommend_games(preds_df,'043a9591-b293-42a8-8d5b-c9ca7f046b3b', games_list, ratings_data)
```
