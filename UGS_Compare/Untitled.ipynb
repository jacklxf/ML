```python
import numpy as np
import pandas as pd
```

```python
ourea=pd.read_csv('C://Users//xiaofeng.li.AFTG//Documents//ML//UGS_Compare//ourea.csv')
ourea.head()
```

```python
ourea=ourea[['Login Name','Total Win/Loss']]
ourea.head()
```

```python
ourea_group=ourea.groupby(['Login Name']).sum().reset_index()
ourea_group['Total Win/Loss']=round(ourea_group['Total Win/Loss'],2)
ourea_group.head()
```

```python
ugs=pd.read_csv('C://Users//xiaofeng.li.AFTG//Documents//ML//UGS_Compare//ugs.csv')
ugs.head()
```

```python
ugs=ugs[['Player','Jackpot Contribution','Company Win/Loss']]
ugs.head()
```

```python
df=pd.merge(ourea_group,ugs,left_on='Login Name',right_on='Player',how='outer')
df=df.fillna(0)
df.isnull().any()
```

```python
df.to_csv('C://Users//xiaofeng.li.AFTG//Documents//ML//UGS_Compare//result.csv')
```
